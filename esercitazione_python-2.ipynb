```python
from abc import ABC, abstractmethod

class Person(ABC):
    def __init__(self, nome, cognome, eta):
        self.nome = nome
        self.cognome = cognome
        self.eta = eta
    
    def saluta(self):
        print(f"Ciao, sono {self.nome} {self.cognome} e ho {self.eta} anni")
    @abstractmethod
    def descrivi(self):
        pass

class Dipendente(Person):
    def __init__(self, nome, cognome, eta, stipendio):
        super().__init__(nome, cognome, eta)
        self.stipendio = stipendio

    def descrivi(self):
        print(f"Ciao, sono {self.nome}, sto lavorando e guadagno {self.stipendio}$ al mese")

class Docente(Dipendente):
    def __init__(self, nome, cognome, eta, stipendio, materia):
        super().__init__(nome, cognome, eta, stipendio)
        self.materia = materia
    def descrivi(self):
        print(f"Ciao, sono {self.nome}, sto insegnando {self.materia} e guadagno {self.stipendio}$ al mese")

class Freelance(Dipendente):
    def __init__(self, nome, cognome, eta, stipendio, partita_iva):
        super().__init__(nome, cognome, eta, stipendio)
        self.partita_iva = partita_iva
    
    def descrivi(self):
        print(f"Ciao, sono {self.nome}, emetto fattura con P.IVA {self.partita_iva} e guadagno circa {self.stipendio}$ al mese")


francesco = Freelance("Francesco", "Scaccetti", 30, 2500, "IT01234567890")
francesco.saluta()
francesco.descrivi()

mario = Docente("Mario", "Rossi", 45, 1800, "Filosofia")
mario.saluta()
mario.descrivi()
```
